## Explanatory Data Analysis 1: Correlation Between Start Order and Finish Order in Races 

```python
# EDA 1: CORRELATION BETWEEN START ORDER AND FINISH ORDER IN RACES

# add dates on the results file to only account for races from 2013 to present
results = pd.read_csv('results.csv')
races = pd.read_csv('races.csv')
recent_race = races[races['year'] >= 2013]
recent_results = pd.merge(recent_race, results, on='raceId')

# clean data to remove DNF's from DF
recent_results['position'] = pd.to_numeric(recent_results['position'], errors='coerce')
race_results = recent_results.dropna(subset='position')



# descriptive statistics
start_stats = race_results['grid'].describe()
finish_stats = race_results['position'].describe()

# correlation

race_results_correlation = race_results['grid'].corr(race_results['position'])

plt.figure(figsize=(15, 6))
sns.regplot(x=race_results['grid'], y=race_results['position'], line_kws={"color":"r","alpha":0.7,"lw":2})
plt.title('Correlation between Starting and Finishing Positions in F1 Races')
plt.xlabel('Starting Position')
plt.ylabel('Finishing Position')
plt.show()

start_stats, finish_stats, race_results_correlation
```

### Analysis
For this data analysis, the graph for this EDA is positively correlated between both the starting and finishing positions in F1. This graph proves that having a better starting position means better chances to have more than a favorable finish. Due to its widespread distribution indicates other factors such as race strategy, driver skillset and the like does heavily significant roles. Not only this figure can be applied by race teams in formulating effective strategies, but also it aims to assess the driver's performance and a tool to attract external sponsors. Upon showing the said statistics to its fans and game analysts, this would be beneficial for them to predict or pick their bet on who would be leading the competition. Hence, this set of data provides a wide array of possible outcomes for past and future races. A regression plot is used for this analysis at it tries to quantify the relationships of both the starting and finishing position, and how it affects the racer's momentum in the tracks.

## Exploratory Data Analysis 2: Qualifiers vs Race Pace

```python
# Exploratory Data Analysis (EDA) 2

## Analysis of Podium Finishes by F1 Drivers in the Last 10 Years

drivers = pd.read_csv('drivers.csv')
drivers_results = pd.merge(recent_results, drivers, on='driverId', how='inner')
podium_results = drivers_results[drivers_results['position'] <= 3]
podium_frequency = podium_results['driverRef'].value_counts()
podium_frequency = podium_frequency.reset_index()
podium_frequency.columns = ['Driver', 'Podium Finishes']

plt.figure(figsize=(15, 8))
sns.barplot(x='Podium Finishes', y='Driver', data=podium_frequency.head(10), palette='rocket')

plt.title('Top 10 F1 Drivers Based on Podium Finishes in the Last 10 Years', fontsize=16)
plt.xlabel('Number of Podium Finishes', fontsize=14)
plt.ylabel('Driver', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

### Analysis
As for this figure, the graph for this EDA is negatively skewed while "hamilton" stays on top of the podium by significantly having the most number of podium finishes for the past decade, as its fellow competitors stay on the bottom having a significant gap among one another. In additon to this, this graph is mostly used but not limited to evaluate the success of each driver in terms of their placement in the podium, and also to keep fans enthused in terms of who is their favorite driver on the racetracks. Factors such as the amount of non-finishes from the races, cars or even the driver themselves could also impact the results of this dataset as neither did not indicate if why drivers not shown on podium due to accidents, or lack of performance of either the car or the driver itself. For this model, the bar graph is appropriate for this comparsion as it only comnpares the amount of podium finishes by each racer.

## Exploratory Data Analysis 3: Overtakeability of a track

```python
import pandas as pd

# Read the 'races.csv' file
races = pd.read_csv('races.csv')

# Read the 'lap_times.csv' file
lap_times = pd.read_csv('lap_times.csv')

# Merge the 'races' and 'lap_times' DataFrames on the 'raceId' column
merged_data = pd.merge(races, lap_times, on='raceId', how='inner')

# Perform your EDA on the merged data
# Calculate the position change between consecutive laps
merged_data['position_change'] = merged_data.groupby(['raceId', 'driverId'])['position'].diff()

# Calculate the number of overtakes (positive position changes) per race
overtakes_per_race = merged_data[merged_data['position_change'] > 0].groupby(['circuitId', 'raceId']).size().reset_index(name='Overtakes')

# Calculate the total overtakes per circuit/track
total_overtakes_per_circuit = overtakes_per_race.groupby('circuitId')['Overtakes'].sum().reset_index()

# Create a bar graph to visualize the total overtakes
plt.figure(figsize=(10, 6))
plt.bar(total_overtakes_per_circuit['circuitId'], total_overtakes_per_circuit['Overtakes'])
plt.xlabel('Circuit/Track ID')
plt.ylabel('Total Overtakes')
plt.title('Total Overtakes per Circuit/Track')
plt.show()
```

## Explanatory Data Analysis 4: Starting Lap Analysis

```python
# EDA 4: Top 10 Car Constructors Based on total wins and average points per season over the past ten years
constructors_data = pd.read_csv('constructors.csv')
constructor_standings = pd.read_csv('constructor_standings.csv')

# Step 1: Determine the range of seasons
latest_season = constructor_standings['raceId'].max()  # assuming raceId represents the season
ten_years_ago = latest_season - 10

# Step 2: Filter records from the last decade
recent_standings = constructor_standings[constructor_standings['raceId'] > ten_years_ago]

# Step 3: Calculate total points for each constructor per season
total_points_per_season = recent_standings.groupby(['constructorId', 'raceId'])['points'].sum().reset_index()

# Step 4: Calculate the average points per season for each constructor
average_points = total_points_per_season.groupby('constructorId')['points'].mean().reset_index()

# Step 5: Sorting and selecting the top 10 constructors
top_10_constructors = average_points.nlargest(10, 'points')

# Merging to get the names
top_10_with_names = pd.merge(top_10_constructors, constructors_data, on='constructorId', how='left')

# Selecting relevant columns to display
top_10_constructors_names_and_points = top_10_with_names[['name', 'points']]

plt.figure(figsize=(15, 8))
sns.barplot(x='points', y='name', data=top_10_constructors_names_and_points, palette='rocket')  # corrected here

plt.title('Top 10 Car Constructors Based on Average Points Per Season Over the Past 10 Years', fontsize=16)
plt.xlabel('Average Pointsz', fontsize=14)
plt.ylabel('Constructor Name', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotating x-labels for better readability
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # It's better to have horizontal grid lines for a bar plot

# Show the plot
plt.tight_layout()
plt.show()
```

### Analysis
The figure represents that it is negatively skewed due to how Red Bull consistently wins races and averages significantly higher than its successors, namely Mercedes and Aston Martin. Given the figure, it is safe to say that it can be used in formulating new strategic decisions that may impact its investments, driver preference and brand sponsorships. Also, it is a quick glance of how a constructor had become dominant within the scene and also allows stakeholders to monitor the performance of every team throughout the entire decade. Since this figure only represents the average score per constructor, it also lacks the context of how a constructor performs when being put into situations such as financial budgeting, stagnating performances, technological competency and more. For this model, it is easier described through a bar graph as it only compares the number of wins car constructors had during the past decade.
