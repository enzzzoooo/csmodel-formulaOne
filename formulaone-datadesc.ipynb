It's common sense to think that the further ahead you start in a race, the better your chances are of finishing ahead of the pack. This analysis aims to quantify the strength of this notion, and understand how much do starting positions influence race outcomes.

```python
# EDA 1: CORRELATION BETWEEN START ORDER AND FINISH ORDER IN RACES

# add dates on the results file to only account for races from 2013 to present
results = pd.read_csv('results.csv')
races = pd.read_csv('races.csv')
recent_race = races[races['year'] >= 2013]
recent_results = pd.merge(recent_race, results, on='raceId')

# clean data to remove DNF's from dataframe
recent_results['position'] = pd.to_numeric(recent_results['position'], errors='coerce')
race_results = recent_results.dropna(subset='position')

# correlation function
race_results_correlation = race_results['grid'].corr(race_results['position'])

race_results_correlation
```

### Analysis

The correlation is at 0.71, signifying a strong positive correlation between the starting and finishing positions in any given race. This, however, does not take into account for driver retirements or vehicular failure/breakage during a race; there are many a time when cars leading a race will suddenly fall behind the pack due to collisions with other drivers, as well as faulty cars and driver error.


```python
plt.figure(figsize=(15, 6))
sns.regplot(x=race_results['grid'], y=race_results['position'], line_kws={"color":"r","alpha":0.7,"lw":2})
plt.title('Correlation between Starting and Finishing Positions in F1 Races')
plt.xlabel('Starting Position')
plt.ylabel('Finishing Position')
plt.show()
```

### Visualization and Interpretation

This regression plot is used for the analysis as it quantifies both the starting and finishing position of each player, and based on the trend it is positively correlated. It also further proves the point that having a head start on the lap does create a significant impact upon racing towards the finish line. External factors such as driver skillset, race strategy and techniques do also make a difference in attaining a higher position in finishing a race. 

## Exploratory Data Analysis 2: Drivers who Frequent the Podium

```python
# Exploratory Data Analysis (EDA) 2

## Analysis of Podium Finishes by F1 Drivers in the Last 10 Years

drivers = pd.read_csv('drivers.csv')
drivers_results = pd.merge(recent_results, drivers, on='driverId', how='inner')
podium_results = drivers_results[drivers_results['position'] <= 3]
podium_frequency = podium_results['driverRef'].value_counts()
podium_frequency = podium_frequency.reset_index()
podium_frequency = podium_frequency.merge(drivers[['driverRef', 'surname', 'forename']], on='driverRef')
podium_frequency['Full Name'] = podium_frequency['forename'].str.cat(podium_frequency['surname'], sep=' ')
podium_frequency.columns = ['Driver', 'Podium Finishes', 'Surname', 'Forename', 'Full Name']
podium_frequency = podium_frequency.drop(columns=['Driver', 'Surname', 'Forename'])
podium_frequency = podium_frequency[['Full Name', 'Podium Finishes']]
podium_frequency.index += 1
podium_frequency.head(10)



```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x='Podium Finishes', y='Full Name', data=podium_frequency.head(10), palette='rocket')

plt.title('Top 10 F1 Drivers Based on Podium Finishes in the Last 10 Years', fontsize=16)
plt.xlabel('Number of Podium Finishes', fontsize=14)
plt.ylabel('Driver', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

### Analysis
As for this figure, the graph for this EDA is negatively skewed while "hamilton" stays on top of the podium by significantly having the most number of podium finishes for the past decade, as its fellow competitors stay on the bottom having a significant gap among one another. In additon to this, this graph is mostly used but not limited to evaluate the success of each driver in terms of their placement in the podium, and also to keep fans enthused in terms of who is their favorite driver on the racetracks. ## Exploratory Data Analysis 3: Overtakeability of a track

```python
# EDA 3: Overtakeability of a track

recent_race_info = recent_race[['raceId', 'circuitId', 'year', 'name', ]].sort_values(by='raceId')

circuit = pd.read_csv('circuits.csv')
lap_times = pd.read_csv('lap_times.csv')
recent_lap_times = lap_times[lap_times['raceId'].isin(recent_race_info['raceId'])]


sorted_lap_times = recent_lap_times.sort_values(by=['raceId', 'driverId', 'lap'])


sorted_lap_times['position_next_lap'] = sorted_lap_times.groupby(['raceId', 'driverId'])['position'].shift(-1)
sorted_lap_times['position_change'] = sorted_lap_times['position_next_lap'] - sorted_lap_times['position']
sorted_lap_times_lastlap = sorted_lap_times[sorted_lap_times['position_next_lap'].notna()]


race_overtakes = sorted_lap_times_lastlap.groupby('raceId')['position_change'].apply(lambda x: x.abs().sum())
race_overtakes = race_overtakes.reset_index(name='total_overtakes')

race_overtakes = pd.merge(recent_race_info, race_overtakes, on='raceId')
race_overtakes_circuit = race_overtakes.groupby('circuitId')['total_overtakes'].sum().reset_index()
race_overtakes_circuitCount = race_overtakes.groupby('circuitId')['raceId'].nunique().reset_index(name='num_races')

overtake_average = pd.merge(race_overtakes_circuit, race_overtakes_circuitCount, on='circuitId')
overtake_average['average'] = overtake_average['total_overtakes'] / overtake_average['num_races']

#race_overtakes = race_overtakes.sort_values(by='total_overtakes', ascending=False)
overtake_average = pd.merge(overtake_average, circuit, on='circuitId')
overtake_average = overtake_average.sort_values(by='average', ascending=False)

track_charts = pd.concat([overtake_average.head(5), overtake_average.tail(5)])

colors = ['green'] * 5 + ['red'] * 5

# Creating the combined bar chart
plt.figure(figsize=(12, 8))
bars = plt.barh(track_charts['name'], track_charts['average'], color=colors)
plt.xlabel('Average Overtakes', fontsize=14)
plt.ylabel('Circuit', fontsize=14)
plt.title('Top 5 vs Bottom 5 F1 Circuits by Average Overtakes (Last 10 Years)', fontsize=16)
plt.gca().invert_yaxis()  # Highest averages on top

plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()






```

```python
# EDA 4: Top 10 Car Constructors Based on total wins and average points per season over the past ten years

# we'll need recent_race

constructors = pd.read_csv('constructors.csv')
constructor_standings = pd.read_csv('constructor_standings.csv')

recent_constructor_standings = pd.merge(recent_race, constructor_standings, on='raceId')
constructor_standings = recent_constructor_standings.merge(constructors, on='constructorId', how='left')

winners = recent_results.loc[recent_results['positionOrder'] == 1]
constructor_wins = winners['constructorId'].value_counts().reset_index()
constructor_wins.columns = ['constructorId', 'wins']
constructor_wins = constructor_wins.merge(constructors[['name', 'constructorId']], on='constructorId')


season_standings = recent_constructor_standings.merge(recent_race[['raceId']], on='raceId')
season_finale = season_standings.groupby(['year', 'constructorId'])['raceId'].max().reset_index()
season_points = pd.merge(season_finale, season_standings, on=['year', 'constructorId', 'raceId'], how='left')
ave_season_points = season_points.groupby('constructorId')['points'].mean().reset_index()
ave_season_points = ave_season_points.merge(constructors[['name', 'constructorId']], on='constructorId')
ave_season_points = ave_season_points.sort_values(by='points', ascending=False)

#season_standings = season_standings.groupby(['constructorId', 'name', 'year']) \
#  .agg(total_wins=('wins', 'sum'), total_points=('points', 'sum')).reset_index()

plt.figure(figsize=(15, 12))

# Horizontal bar chart for wins
plt.subplot(2, 1, 1)  # 2 rows, 1 column, first plot
plt.barh(constructor_wins['name'].head(10), constructor_wins['wins'].head(10), color='skyblue')
plt.ylabel('Constructors')
plt.xlabel('Number of Wins')
plt.title('Top 8 Constructors by Number of Wins (Last 10 Years)')
plt.gca().invert_yaxis()  # Invert y-axis to have the constructor with the most wins at the top

# Horizontal bar chart for average points per season
plt.subplot(2, 1, 2)  # 2 rows, 1 column, second plot
plt.barh(ave_season_points['name'].head(10), ave_season_points['points'].head(10), color='lightgreen')
plt.ylabel('Constructors')
plt.xlabel('Average Points per Season')
plt.title('Top 10 Constructors by Average Points per Season (Last 10 Years)')
plt.gca().invert_yaxis()  # Invert y-axis to have the constructor with the most points at the top

# Show the visualizations
plt.show()
```
